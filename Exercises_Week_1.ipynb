Next, let's look into some text data. We will be looking into Amazon reviews, and the necessary steps to transform a raw dataset into a format more suitable for prediction tasks.

2. Next, let's follow some steps to normalize the text data.

When dealing with natural language, it is important to notice that while, for example, the words "Copper" and "copper" are represented by two different strings, they have the same meaning. When applying statistical methods on this data, it is useful to ensure that words with the same meaning are represented by the same string.

* <span style="background-color: #ef0303ff">Downcasing</span>: Let's first downcase the contents of the <span style="background-color: #e90a24ff">reviewText</span> field.

Now the first review should be:

*'after i wrote the below review, the manufacturer contacted me and explained how to use this.  instead of the (current) picture on amazon where the phone is placed vertically, you actually use the stand with the phone placed horizontally.'*

```python
automotive['reviewText']=automotive['reviewText'].str.lower()

automotive

#Not sure if this is computationally the fastest
```

3. Let's continue with punctuation and stop word removal. Stop words are words like "and", "the", etc. They are usually very common words that have little to do with the actual content matter. There's plenty openly available lists of stop words for almost any (natural) language.

* <span style="background-color: #f90d19ff">Punctuation and stop-word removal</span>: Let's now remove all punctuation, as well as the stop words. You can find a stop word list for English, e.g. [here](https://gist.github.com/xldrkp/4a3b1a33f10d37bedbe0068f2b4482e8#file-stopwords-en-txt).*(use the link to download a txt of english stopwords)* Save the stopwords in the file as "stopwords-en.txt".

First review at this point reads as: 

*'wrote review manufacturer contacted explained current picture amazon phone vertically stand phone horizontally'*

```python
import string
stop_words = set(open("stopwords-en.txt").read().split('\n'))

def sentence_parser(sentence):
    sentence = str(sentence)
    r_sentence = []
    sentence = sentence.translate(str.maketrans('', '', string.punctuation))
    for x in sentence.split():
        if x not in stop_words:
            r_sentence.append(x)
    result = ' '.join(r_sentence)
    return result

# Again not sure if this is the fastest


for row in range(len(automotive)):
    automotive.loc[row, 'reviewText'] = sentence_parser(automotive.loc[row, 'reviewText'])


# Well this isin't matching to the one above but the stop word list doesn't contain for example the words like use, instead, placed etc so I'm pretty sure this is correct.
```

4. Let's continue with stemming. For example, while the words "swims" and "swim" are different strings, they both refer to swimming. [Stemming](https://en.wikipedia.org/wiki/Stemming) refers to the process of mapping words from their inflected form to their base form, for instance: swims -> swim.

* <span style="background-color: #f11316ff">Stemming</span>: Apply a stemmer on the paragraphs, so that inflected forms are mapped to the base form. For example, for Python the popular natural language toolkit [nltk](http://www.nltk.org/howto/stem.html) has an easy to use stemmer. In case you are using R, you can try the [Snowball stemmer](https://www.rdocumentation.org/packages/corpus/versions/0.10.2/topics/stem_snowball). You can find out how to install nltk from [here](https://www.nltk.org/install.html). It will take a while to run! So, grab a coffee and wait :D

Finally, after stemming: 

*'wrote review manufactur contact explain current pictur amazon phone vertic stand phone horizont'*

```python
# Use this cell for your code
stemmer = nltk.PorterStemmer()

def stem(sentence):
    r_sentence=[]
    for x in sentence.split():
        r_sentence.append(stemmer.stem(x))
    result = ' '.join(r_sentence)
    return result
        

for row in range(len(automotive)):
    automotive.loc[row, 'reviewText'] = stem(automotive.loc[row, 'reviewText'])
```

5. Finally, filter the data by selecting reviews where the field <span style="background-color: #f6080cff">overall</span> is 4 or 5, and store the review texts in a file named <span style="background-color: #f9162dff">pos.txt</span>. Similarly, select reviews with rating 1 or 2 and store them in a file named <span style="background-color: #f90c1cff">neg.txt</span>. Ignore the reviews with overall rating 3. Each line in the two files should contain exactly one preprocessed review text without the rating.

```python
# Use this cell for your code
pos = automotive[automotive['overall']>3]['reviewText']

neq = automotive[automotive['overall']<3]['reviewText']

pos.to_csv('pos.txt')
neq.to_csv('neq.txt')

```
